```python
# Import case_data
import_path = os.path.join(RAW_DATA_DIR, 'data_extraction', 'case_report_data_extraction_end.xlsx')
case_data = pd.read_excel(import_path)
```

```python
case_data.head()
```

```python
# Check Patient_IDs are unique
case_data['Patient_ID'].nunique() == len(case_data)
```

```python
# Check Study_ID matches Patient_ID
case_data['Study_ID'] = case_data['Study_ID'].astype(str)
(case_data['Study_ID'] != case_data['Patient_ID'].str.split("-").str[0]).sum()
```

```python
case_data[case_data['Study_ID'] != case_data['Patient_ID'].str.split("-").str[0]]
```

```python
non_binary_cols = ['Study_ID', 'Patient_ID', 'Age_Yrs', 'Gender', 'Comments']
binary_cols = [col for col in case_data.columns if col not in non_binary_cols and not col.endswith('Long')]
binary_cols
```

```python
# Check for blank binary_cols
filtered_case_data = case_data[case_data[binary_cols].isna().any(axis=1)]
filtered_case_data
```

```python
# Check for blank age column
filtered_case_data = case_data[case_data['Age_Yrs'].isna()]
filtered_case_data
```

```python
# Check for blank gender column
filtered_case_data = case_data[case_data['Gender'].isna()]
filtered_case_data
```

```python
permitted_values = ['UK', 'Male', 'Female']

# Create a mask where any value in Gender col is not in permitted_values
mask = ~case_data['Gender'].isin(permitted_values)

# Filter the rows that have problems
filtered_case_data = case_data[mask]
filtered_case_data
```

```python
# Replace 'UK' with np.nan 
case_data['Age_Yrs'] = case_data['Age_Yrs'].replace('UK', np.nan)

# convert to float
case_data['Age_Yrs'] = case_data['Age_Yrs'].astype(float)

# check for unrealistic ages
unrealistic_ages = case_data[(case_data['Age_Yrs'] < 0) | (case_data['Age_Yrs'] > 120)]

unrealistic_ages
```

```python
# Check for blank 'Study_ID'
filtered_case_data = case_data[case_data['Study_ID'].isna()]
filtered_case_data
```

```python
# Check for blank 'Patient_ID'
filtered_case_data = case_data[case_data['Patient_ID'].isna()]
filtered_case_data
```

```python
# Check for invalid values in binary cols
permitted_values = ['Y', 'N', 'UK']

# Create a mask where any value across binary_cols is not in permitted_values
mask = ~case_data[binary_cols].isin(permitted_values).all(axis=1)

# Filter the rows that have problems
filtered_case_data = case_data[mask]
filtered_case_data
```

### Filter Dataset

```python
# filter out cases where motivation is unknown
filtered_case_data = case_data[case_data['Motivation_Unknown'] != 'Y']
print(f"Original rows: {len(case_data)}, Filtered rows: {len(filtered_case_data)}")
case_data = filtered_case_data.copy()
```

```python
# Import full text screen end data to see which papers are included
import_path = os.path.join(PROC_DATA_DIR, 'full_text_screen', 'full_text_included_final.csv')
ft_included_data = pd.read_csv(import_path)
ft_included_data['Study_ID'] = ft_included_data['Study_ID'].astype(str)
ft_included_ids = list(ft_included_data['Study_ID'])
```

```python
# Import bibliography search screening results
import_path = os.path.join(PROC_DATA_DIR, 'bibliography_search', 'bib_search_included_final.csv')
bs_included_data = pd.read_csv(import_path)
bs_included_data['Study_ID'] = bs_included_data['Study_ID'].astype(str)
bs_included_ids = list(bs_included_data['Study_ID'])
```

```python
# Create list of all included_ids
included_ids = sorted(ft_included_ids + bs_included_ids)
```

```python
len(included_ids)
```

```python
filtered_case_data = case_data[case_data['Study_ID'].isin(included_ids)]
print(f"Original rows: {len(case_data)}, Filtered rows: {len(filtered_case_data)}")
case_data = filtered_case_data.copy()
```
