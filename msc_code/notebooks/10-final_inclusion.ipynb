```python
# Import libraries
from msc_code.scripts.notebook_setup import *
```

```python
# Import full text screening inclusion data
import_path = os.path.join(PROC_DATA_DIR, 'full_text_screen', 'full_text_screen_included_final.csv')
ft_included_data = pd.read_csv(import_path)
```

```python
# Import included papers from bibliography search
import_path = os.path.join(PROC_DATA_DIR, 'bibliography_search', 'bib_search_included_final.csv')
bs_included_data = pd.read_csv(import_path)
```

```python
# Define desired columns 
desired_columns = ['Study_ID', 'Publication_Year', 'Authors', 'Title', 'Publication_Title', 'DOI']
# View columns 
ft_included_data.columns
```

```python
# Import paper_info to merge 
import_path = os.path.join(PROC_DATA_DIR, 'title_abstract_review', 'ta_review_clean.csv')
paper_info = pd.read_csv(import_path)
paper_info.columns
```

```python
# Create list of included IDs
ft_included_ids = list(ft_included_data['Study_ID'])
```

```python
# Filter to only include data for included studies
mask = paper_info['Study_ID'].isin(ft_included_ids)
paper_info = paper_info[mask]
```

```python
# Rename columns
paper_info.columns = paper_info.columns.str.strip().str.replace(" ", "_")
paper_info.columns
```

```python
# Extract desired information
paper_info = paper_info[desired_columns]
```

```python
# Rename Author column, as contains multiple authors.
bs_included_data = bs_included_data.rename(columns={
    'Author': 'Authors'
})
bs_included_data.columns
```

```python
# Extract Desired Columns
bs_included_data = bs_included_data[desired_columns]
```

```python
# Concatinate all data to produce final
included_paper_info = pd.concat(
    [paper_info, 
    bs_included_data]
)

# Sort by Study_ID ascending
included_paper_info = included_paper_info.sort_values(by='Study_ID', ascending=True)
```

```python
# Export to CSV
export_path = os.path.join(PROC_DATA_DIR, 'final_inclusion', 'included_paper_data.csv')
included_paper_info.to_csv(export_path, index=False)
```
